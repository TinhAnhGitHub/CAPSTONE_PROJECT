```python
from typing import Callable, Dict, List, Optional, Any, get_type_hints
import inspect
from pydantic import BaseModel, Field
from llama_index.core.tools import FunctionTool
from dataclasses import dataclass


@dataclass(frozen=True)
class ToolMetadata:
    """Metadata extracted from tool function"""
    func: Callable
    category: str | None = None
    tags: List[str] | None = None
    dependencies: List[str] | None = None

    @property
    def name(self) -> str:
        """Function name"""
        return self.func.__name__
    
    @property
    def signature(self) -> inspect.Signature:
        """Function signature"""
        return inspect.signature(self.func)

    @property
    def docstring(self) -> str | None:
        """Function docstring"""
        return inspect.getdoc(self.func)
    
    @property
    def type_hints(self) -> dict[str, Any]:
        """Function type hints"""
        return get_type_hints(self.func)

    
    @property
    def output_type(self) -> Optional[Any]:
        """Extracted return type (output type) from function annotation"""
        type_hints = self.type_hints.get('return', None)
```

```python
class Output(BaseModel):
    response: str = Field(..., description="Hello")


async def heloo() -> list[Output]:
    return Output(response="Hi")
```

```python
tool_metadata = ToolMetadata(
    func=heloo,
    category="hi",
    tags=["hi"],
    dependencies=['hi']
)
```
