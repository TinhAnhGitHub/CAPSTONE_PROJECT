```python
from dotenv import load_dotenv
from google.genai import types
load_dotenv()
```

```python
generation_config = types.GenerateContentConfig(
    temperature=0.7,
    thinking_config=types.ThinkingConfig(thinking_budget=1024, include_thoughts=True)
)
```

```python
from llama_index.llms.google_genai import GoogleGenAI
from llama_index.core.base.llms.types import ThinkingBlock, ContentBlock, TextBlock
from llama_index.core.llms import ChatMessage, ChatResponse, MessageRole

llm = GoogleGenAI(model="gemini-2.5-pro", generation_config=generation_config)

prompt = ChatMessage(
   role=MessageRole.USER,
   content="Given 5 distinct numbers, how many ways can we rearrange them?" 
)


thought = ""
last_len= 0
async for r in await llm.astream_chat([prompt]):
    message_blocks: list[ContentBlock] = r.message.blocks
    thinking_blocks = list(filter(lambda x: isinstance(x, ThinkingBlock), message_blocks))
    text_block: TextBlock = next(filter(lambda x: isinstance(x, TextBlock), message_blocks), None) #type:ignore
    
    if text_block:
        text_spawn = text_block.text
        new_text = text_spawn[last_len:]
        if new_text:
            print(new_text, end="", flush=True)
        last_len = len(text_spawn)
       
        continue

    print(thinking_blocks[-1])
    
```

```python
from videodeepsearch.agent.custom import StreamingFunctionAgent
from llama_index.core.tools import FunctionTool


def add(x,y) -> int| float:
    """
    use this tool for addition
    """
    return x + y

def minus(x,y) -> int| float:
    """
    use this tool for minus
    """
    return x - y

def mul(x,y) -> int| float:
    """
    use this tool for multiply
    """
    return x * y

def div(x,y) -> int| float:
    """
    use this tool for addition
    """
    
    return x / y

agent = StreamingFunctionAgent(
    name='Math agent',
    tools=[
        FunctionTool.from_defaults(add),
        FunctionTool.from_defaults(minus),
        FunctionTool.from_defaults(div),
        FunctionTool.from_defaults(mul)
    ],
    llm=llm,
    streaming=True        
)
```

```python
handler = agent.run("What is 1 + 1 * 312313/ 22332 + 3229 * (0009833) ? Think carefully before answer, and use tools")
async for event in handler.stream_events():
    print()
    print(type(event))
    print(event)
    print()
answer = await handler
```
