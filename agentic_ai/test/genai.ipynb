```python
from dotenv import load_dotenv
from google.genai import types
load_dotenv()
```

```python
generation_config = types.GenerateContentConfig(
    temperature=0.7,
    thinking_config=types.ThinkingConfig(thinking_budget=1024, include_thoughts=True)
)
```

```python
from llama_index.llms.google_genai import GoogleGenAI

llm = GoogleGenAI(model="gemini-2.5-pro", generation_config=generation_config)


seen_thoughts = ""
for r in llm.stream_complete("Given 5 distinct numbers, how many ways can we rearrange them?"):
    # new reasoning tokens (diff-based)
    thoughts = r.additional_kwargs.get("thoughts", "")
    if thoughts and thoughts != seen_thoughts:
        new_part = thoughts[len(seen_thoughts):]
        print("🧠 THINKING:", new_part, end="")
        seen_thoughts = thoughts
        continue

    # print answer tokens (no filtering by thoughts)
    if r.delta:
        print(r.delta, end="")
```
